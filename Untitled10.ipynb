```python
import pandas as pd
import numpy as np
from scipy import stats


def website_analyse(file_path):
  df = pd.read_csv(file_path)

  # Среднее кол-во просмотренных страниц
  mean_pages = df['Просмотренные страницы'].mean()
  print(f"1) Среднее количество просмотренных страниц: {mean_pages:.2f}")

  # Стандартное отклонение времени на сайте
  std_time = df['Время на сайте (сек)'].std()
  print(f"2) Стандартное отклонение: {std_time:.2f}")

  # Коэффициент вариации
  mean_time = df['Время на сайте (сек)'].mean()
  cv = std_time / mean_time

  if cv > 0.25:
    Вариабельность = "Высокая вариабельность"
  elif cv > 0.1:
    Вариабельность = "Умеренная вариабельность"
  else:
    Вариабельность = "Слабая вариабельность"

  print(f"3) Вариабельность: {cv:.2f}")

  # Проверка статистической значимости
  mobile_data = df[df['Тип устройства'] == 'Мобильное']['Просмотренные страницы']
  desktop_data = df[df['Тип устройства'] == 'Десктоп']['Просмотренные страницы']

  t_stat, p_value = stats.ttest_ind(mobile_data, desktop_data)
  if p_value > 0.5:
    result = "Нет статистически значимой разницы"
  else:
    result = "Есть статистически значимая разница в количестве страниц"

  print(f"4) {result}: {p_value:.4f}")

website_analyse('/content/website_visits_december_2024 (1).csv')
```
